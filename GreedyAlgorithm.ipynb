Define function and variable

```python
def dijkstra(matrix, start_node):     
    '''
   #matrix: (traffic matric) representing the cities in the network and the distances between them  
   #start_node：The number of the starting site (starting from 0 to 18, respectively), 
                indicating which site to start looking for the shortest path from  
    '''
    num_nodes = len(matrix)      
    '''
    The total number of nodes (the number of cities in the network) is 
    obtained through len(graph) and is used to traverse in the loop.
    '''
    visited = [False] * num_nodes  
    #Create a list where all variables are Boolean values ​​and the number of variables is equal to the number of nodes.
    infinity = float('inf')       
    shortest_paths = [infinity ] * num_nodes 
    '''
    Create a list named shortest_paths, which is similar to P (Waterloo) in the Working Example, to record the distance 
    from the starting point to the end point. P(P) is initialized to infinity.
    '''
    shortest_paths [start_node] = 0 
    #The start distance is 0.
    previous_nodes =[-1] * num_nodes  
    #Record the previous node of each node. The initial value is -1 which means no predecessor node.
    #Write a loop to find the shortest_paths, previous_nodes (shortest path, previous node)
    for _ in range(num_nodes):  #traverse
    # Initialize the current shortest distance and the index of the currently found closest node.
        min_distance = infinity
        current_node = -1
    #Find new unvisited points（shortest）
        for node in range(num_nodes):
            if not visited[node] and shortest_paths[node] < min_distance:
                min_distance = shortest_paths[node]
                current_node = node
        # End function if it is not found
        if current_node == -1:
            break
        # Mark new city as visited
        visited[current_node] = True

        # Update the distance and record provious node
        for neighbor in range(num_nodes):
            if matrix[current_node][neighbor] > 0 and not visited[neighbor]:
                new_distance = shortest_paths[current_node] + matrix[current_node][neighbor]
                if new_distance < shortest_paths[neighbor]:
                    shortest_paths[neighbor] = new_distance   #The shortest distance from the starting point to each node
                    previous_nodes[neighbor] = current_node   #Previous node

    return shortest_paths, previous_nodes 

    
    
```

```python
# Defining a path function.
def get_path(previous_nodes, start_node, target_node):
    # The starting and ending points are the same, which means that there is no change in position.
    if start_node == target_node:
        print(f"Route: {start_node}")
        return [start_node]
        
    # The starting and ending points are not the same situation.    
    else:
        path = []
        current_node = target_node
        
        while current_node != -1:  #Traced back the starting point.(current_node == -1,means had done.)
            path.append(current_node)       # add previous node
            current_node = previous_nodes[current_node]
        path.reverse()                      #reverse route
        print("Route: ", " → ".join(map(str, path)))
    #That is a clear logic and return path
        return path
```
